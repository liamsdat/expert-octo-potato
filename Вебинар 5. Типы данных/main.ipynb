# 3. Циклы while, for

# While - итерационный цикл
# For - счётный цикл

```python
x = 5
while x != 0:
    x -= 1
    print(x)

print("Цикл завершил все итерации")
```

```python
x = 7
while x != 0:
    if x % 2 == 0:
        print(x, '- чётное число')
    else:
        print(x, '- нечётное число')
    x -= 1
```

```python
# Итерация по строкам
company_name = 'Orange'
for letter in company_name:
    print(letter)
    letter = letter.capitalize() 
    # Метод .capitalize() в Python возвращает копию строки, в которой первый символ преобразован в заглавный (верхний регистр),
    # а все остальные символы — в строчные (нижний регистр).
    # В данном случае он возвращает все символы строки заглавными, т.к считает их за новые строки
    print(letter)
print(letter)
```

```python
companies_capitalization = [
    'Amazon',
    'Microsoft',
    'Mojang',
    'Electronic Arts'
]


for company in companies_capitalization:
    #print(company)
    print(company[0], 'capitalization is', company[1])
    print('end of iteration')
```

```python
# break, pass, continue
phrase = '640кБ должно хватить для любых задач. Билл Гейтс (по легенде)'
```

```python
for letter in phrase:
    if letter == ' ':
        break
    print(letter, end='')
```

```python
for letter in phrase:
    if letter == ' ':
        continue
    print(letter, end='')
print('\nfinish loop')
```

```python
for letter in phrase:
    if letter == ' ':
        pass
    print(letter, end= '')
print('\nfinish loop')
```

```python
cook_book = [
  ['салат',
      [
        ['картофель', 100, 'г'],
        ['морковь', 50, 'г'],
        ['огурцы', 50, 'г'],
        ['горошек', 30, 'г'],
        ['майонез', 70, 'мл'],
      ]
  ],
  ['пицца',
      [
        ['сыр', 50, 'г'],
        ['томаты', 50, 'г'],
        ['тесто', 100, 'г'],
        ['бекон', 30, 'г'],
        ['колбаса', 30, 'г'],
        ['грибы', 20, 'г'],
      ],
  ],
  ['фруктовый десерт',
      [
        ['хурма', 60, 'г'],
        ['киви', 60, 'г'],
        ['творог', 60, 'г'],
        ['сахар', 10, 'г'],
        ['мёд', 50, 'мл'],
      ]
  ]
]
```

```python
person = 5
print(cook_book[0][1])
```

```python
for item in cook_book:
    print(item[0].capitalize() + ':')
    
    for line in item[1]:
        meal, weight, gr = line
        weight = weight * person

        print(' ' + meal, weight, gr)

    print('')
```

```python
enumerate(cook_book)

for item in cook_book:
    print(item[0].capitalize() + ':')
    for elem in item[1]:
        meal = elem[0]
        weight = elem[1]
        gr = elem[2]
        weight = weight * person
        print(" ", meal, weight, gr)
```
