```python
del(salaries['Nick'])
salaries
```

```python
# hard dict
staff_dict = {
    'Alex' : {'salary' : 800, 'bonus': 200},
    'Andrew' : {'salary' : 700, 'bonus': 300},
    'Marcus' : {'salary' : 600, 'bonus': 400},
    'Sitara' : {'salary' : 1100, 'bonus': 900}
}
```

```python
staff_dict['Andrew']['bonus']
```

```python
staff_dict['Jan'] = {'salary' : 1000, 'bonus' : 200}
staff_dict['Jan']['salary']
```

```python
print(staff_dict.keys(), '\n')
print(staff_dict.values(), '\n')
print(staff_dict.items(), '\n')
```

```python
for person in staff_dict.keys():
    print(person)
```

```python
for value in staff_dict.values():
    print(value)
```

```python
for value in staff_dict.values():
    for salary, bonus in value.items():
        print(salary, bonus)
```
