```python
month_list = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

income_list = [13_000, 14_000, 15_000, 13_500, 17_600, 14_300, 15_000, 14_300, 15_000, 12_350, 16_500, 11_400]

income_by_month = [
    ['Jan', 13_000],
    ['Feb', 14_000],
    ['Mar', 15_000],
    ['Apr', 13_500],
    ['May', 17_600],
    ['Jun', 14_300],
    ['Jul', 15_000],
    ['Aug', 14_300],
    ['Sep', 15_000],
    ['Oct', 12_350],
    ['Nov', 16_500],
    ['Dec', 11_400],

]
```

```python
print(type(month_list))
print(type(income_list))
print(type(income_by_month))
```

```python
# индексация элементов в списке
print(month_list[0])
print(month_list[-1])
print(income_by_month[8][1])
```

```python
# срезы
print(income_by_month[1:3])
print(income_by_month[0:-3])
print(income_by_month[0:9])
print(income_by_month[2:5])
```

```python
# Изменение списков
income_by_month[0][1] = 12_900
print(income_by_month)
```

```python
income_by_month[0:3] = [['Jan', 12_900], ['Feb', 12_345], ['Mar', 11_700]]
print(income_by_month)
```

```python
a, b = income_by_month[0:2]
```
